## 2.3 関数の作成

```python
def my_func(in_value):
  out_value = in_value * 10
  return out_value

print(my_func(in_value = 5))
print(my_func(5))
print("out_value =", my_func(5))
print(f"out_value = {my_func(5)}")
```

## 2.6 pandasのDataFrame

```python
# DataFrame
df_1 = pd.DataFrame({
    "column1": [1, 2, 3], 
    "column2": [4, 5, 6]
})
print(df_1)

# インデックス（行名）付与
df_1.index = ["row1", "row2", "row3"]
print(df_1)
```

## 2.7 pandasのSeries

```python
# Series
series_1 = pd.Series([7, 8])
print(series_1)

# インデックス付与
series_1.index = ["idx1", "idx2"]
print(series_1)
```

```python
# df_1の1列目を抽出
series_2 = df_1["column1"]
print(series_2)
```

## 2.8 DataFrameの演算

```python
# 元のdf_1
print(df_1)

# 1を足す
print(df_1 + 1)
print(df_1.add(1))

# 1を引く
print(df_1.sub(1))

# 2をかける
print(df_1.mul(2))

# 2で割る
print(df_1.div(2))
```

## 2.9 DataFrameの行や列を対象とした演算

```python
# 列ごとに最大値を取得
df_1.max()
```

```python
# 行ごとに最大値を取得
df_1.max(axis = 1)
```

## 2.10 DataFrameに対する関数の適用

```python
# 底が2である対数の計算
np.log2(4)
```

```python
# 元のdf_1
print(df_1)

# すべての要素の対数を取る
print(df_1.apply(np.log2))
```

```python
# 列ごとに合計を計算
df_1.apply(np.sum, axis = 0)
```

```python
# 行ごとに合計を計算
print(df_1.apply(np.sum, axis = 1))
print(df_1.sum(axis = 1))
```
